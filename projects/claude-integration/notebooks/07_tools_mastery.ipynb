## File Operation Tools

### Read Tool

Read file contents with line numbers and optional pagination.

**Parameters:**
- `file_path` (required): Absolute path to file
- `offset` (optional): Line number to start from
- `limit` (optional): Number of lines to read

**Examples:**

```python
# Read entire file
Read(file_path="/home/user/project/src/app.py")

# Read lines 100-200
Read(
    file_path="/home/user/project/large_file.log",
    offset=100,
    limit=100
)
```

**Key Features:**
- Shows line numbers for easy reference
- Can read images, PDFs, and Jupyter notebooks
- Handles large files with pagination
- Truncates lines longer than 2000 characters

**Best Practices:**
- Always read files before editing them
- Use offset/limit for very large files
- Read multiple related files in parallel

### Write Tool

Create new files or completely overwrite existing ones.

**Parameters:**
- `file_path` (required): Absolute path to file
- `content` (required): Complete file content

**Examples:**

```python
# Create new file
Write(
    file_path="/home/user/project/src/config.py",
    content="""# Configuration
API_KEY = 'your-key-here'
DEBUG = True
"""
)
```

**Important Rules:**
- Must read existing files before overwriting
- Prefer Edit over Write for existing files
- Always use absolute paths
- Creates parent directories if they don't exist

**When to Use:**
- Creating brand new files
- Generating configuration files
- Creating documentation
- Writing test files

**When NOT to Use:**
- Modifying existing code (use Edit)
- Making small changes (use Edit)
- Updating part of a file (use Edit)

### Edit Tool

Make precise, surgical changes to existing files using exact string replacement.

**Parameters:**
- `file_path` (required): Absolute path to file
- `old_string` (required): Exact text to replace
- `new_string` (required): Replacement text
- `replace_all` (optional): Replace all occurrences (default: false)

**Examples:**

```python
# Single replacement
Edit(
    file_path="/home/user/project/src/app.py",
    old_string="DEBUG = True",
    new_string="DEBUG = False"
)

# Replace all occurrences (rename variable)
Edit(
    file_path="/home/user/project/src/utils.py",
    old_string="old_var_name",
    new_string="new_var_name",
    replace_all=True
)

# Multi-line replacement with exact indentation
Edit(
    file_path="/home/user/project/src/main.py",
    old_string="""def process_data(data):
    return data""",
    new_string="""def process_data(data):
    # Validate input
    if not data:
        raise ValueError(\"Data cannot be empty\")
    return data"""
)
```

**Critical Rules:**
1. Must read file first
2. `old_string` must match EXACTLY (including whitespace)
3. Preserve indentation exactly as shown in Read output
4. Don't include line numbers from Read output
5. If `old_string` appears multiple times, Edit fails (unless using `replace_all`)

**Best Practices:**
- Include surrounding context to make `old_string` unique
- Use `replace_all` for renaming variables/functions
- Test Edit on small changes first
- Preserve exact indentation (tabs vs spaces)

### Glob Tool

Find files matching glob patterns - incredibly fast even on large codebases.

**Parameters:**
- `pattern` (required): Glob pattern to match
- `path` (optional): Directory to search in

**Pattern Syntax:**
- `*` - Matches any characters except `/`
- `**` - Matches any characters including `/`
- `?` - Matches single character
- `[abc]` - Matches one of the characters
- `{a,b}` - Matches either pattern

**Examples:**

```python
# Find all Python files
Glob(pattern="**/*.py")

# Find test files
Glob(pattern="**/*test*.py")

# Find config files
Glob(pattern="**/*.{json,yaml,yml,toml}")

# Find in specific directory
Glob(
    pattern="*.js",
    path="/home/user/project/src"
)

# Find React components
Glob(pattern="**/*.{jsx,tsx}")

# Find markdown docs
Glob(pattern="**/*.md")
```

**Use Cases:**
- Finding all files of a certain type
- Locating test files
- Discovering configuration files
- Building file lists for batch operations

**Pro Tips:**
- Results sorted by modification time (newest first)
- Much faster than `find` command
- Respects .gitignore by default
- Can run multiple Globs in parallel

### Grep Tool

Search file contents using regex patterns - powered by ripgrep for blazing speed.

**Parameters:**
- `pattern` (required): Regex pattern to search
- `path` (optional): Directory/file to search
- `output_mode` (optional): "content", "files_with_matches", or "count"
- `glob` (optional): Filter files by pattern
- `type` (optional): Filter by file type (js, py, rust, etc.)
- `-i` (optional): Case insensitive search
- `-A`, `-B`, `-C` (optional): Context lines after/before/both
- `multiline` (optional): Enable multiline matching
- `head_limit` (optional): Limit number of results

**Examples:**

```python
# Find function definitions
Grep(
    pattern="def \\w+\\(",
    type="py",
    output_mode="content"
)

# Find TODO comments
Grep(
    pattern="TODO|FIXME",
    output_mode="content"
)

# Case-insensitive search
Grep(
    pattern="api_key",
    **{"-i": True},
    output_mode="files_with_matches"
)

# Search with context
Grep(
    pattern="class \\w+",
    **{"-A": 5, "-B": 2},
    output_mode="content"
)

# Search in JavaScript files only
Grep(
    pattern="import .* from",
    glob="*.{js,jsx}",
    output_mode="content"
)

# Count occurrences
Grep(
    pattern="console\\.log",
    type="js",
    output_mode="count"
)

# Multiline search for class definitions
Grep(
    pattern="class \\w+[\\s\\S]*?\\{",
    multiline=True,
    output_mode="content"
)
```

**Output Modes:**
- `files_with_matches`: Just list files (default, fastest)
- `content`: Show matching lines with line numbers
- `count`: Show count of matches per file

**Best Practices:**
- Use `files_with_matches` first to narrow scope
- Add `head_limit` for large result sets
- Use `type` or `glob` to filter file types
- Escape special regex characters: `\\.`, `\\(`, `\\{`
- Use multiline sparingly (slower)

## Execution Tools

### Bash Tool

Execute shell commands in a persistent session.

**Parameters:**
- `command` (required): Shell command to execute
- `description` (optional): Human-readable description
- `timeout` (optional): Timeout in milliseconds (default: 120000, max: 600000)
- `run_in_background` (optional): Run command in background

**Examples:**

```python
# Run tests
Bash(
    command="pytest tests/",
    description="Run pytest test suite"
)

# Git operations (sequential)
Bash(
    command="git add . && git commit -m 'Update feature' && git push",
    description="Stage, commit, and push changes"
)

# Install dependencies
Bash(
    command="npm install",
    description="Install npm dependencies",
    timeout=300000  # 5 minutes
)

# Run development server in background
Bash(
    command="npm run dev",
    description="Start development server",
    run_in_background=True
)

# Handle paths with spaces
Bash(
    command='cd "/path/with spaces" && python script.py',
    description="Run script in directory with spaces"
)
```

**Command Chaining:**
- `&&` - Run next command only if previous succeeds
- `||` - Run next command only if previous fails
- `;` - Run commands sequentially regardless of success
- `&` - Run in background (or use `run_in_background=True`)

**Key Rules:**
- Use absolute paths (cwd resets between calls)
- Quote paths with spaces
- Chain dependent commands with `&&`
- Don't use interactive commands (-i flag)
- Prefer specialized tools over bash:
  - Use Read instead of `cat`
  - Use Glob instead of `find`
  - Use Grep instead of `grep`
  - Use Edit instead of `sed`

### Background Process Management

**Starting Background Processes:**

```python
# Start server
Bash(
    command="python manage.py runserver",
    run_in_background=True
)

# Start file watcher
Bash(
    command="npm run watch",
    run_in_background=True
)
```

**Monitoring Output:**

```python
# Read output from background process
BashOutput(
    bash_id="shell-123",  # Get from /bashes command
)

# Filter output with regex
BashOutput(
    bash_id="shell-123",
    filter="ERROR|WARNING"
)
```

**Stopping Processes:**

```python
# Kill background shell
KillShell(
    shell_id="shell-123"
)
```

**Finding Shell IDs:**
Use the `/bashes` slash command to list all running background shells.

**Use Cases:**
- Development servers
- File watchers
- Long-running builds
- Database servers
- Test runners in watch mode

## Todo Management

### TodoWrite Tool

Track multi-step tasks with a structured todo list.

**When to Use:**
- Complex tasks with 3+ steps
- Multiple user requests
- Non-trivial implementations
- User explicitly requests todo tracking

**When NOT to Use:**
- Single, simple tasks
- Trivial operations
- Purely informational requests

**Task Structure:**

```python
TodoWrite(
    todos=[
        {
            "content": "Create user authentication module",
            "activeForm": "Creating user authentication module",
            "status": "in_progress"
        },
        {
            "content": "Write unit tests for auth",
            "activeForm": "Writing unit tests for auth",
            "status": "pending"
        },
        {
            "content": "Update documentation",
            "activeForm": "Updating documentation",
            "status": "pending"
        }
    ]
)
```

**Task States:**
- `pending`: Not yet started
- `in_progress`: Currently working (only ONE at a time)
- `completed`: Successfully finished

**Critical Rules:**
1. Each task needs `content` AND `activeForm`
2. Only ONE task in_progress at a time
3. Complete tasks immediately when done
4. Only mark completed when FULLY accomplished
5. Update list after each significant step

**Example Workflow:**

```python
# Step 1: Create initial todo list
TodoWrite(
    todos=[
        {"content": "Implement feature X", "activeForm": "Implementing feature X", "status": "in_progress"},
        {"content": "Write tests", "activeForm": "Writing tests", "status": "pending"},
        {"content": "Update docs", "activeForm": "Updating docs", "status": "pending"}
    ]
)

# Step 2: After implementing feature
TodoWrite(
    todos=[
        {"content": "Implement feature X", "activeForm": "Implementing feature X", "status": "completed"},
        {"content": "Write tests", "activeForm": "Writing tests", "status": "in_progress"},
        {"content": "Update docs", "activeForm": "Updating docs", "status": "pending"}
    ]
)

# Step 3: After writing tests
TodoWrite(
    todos=[
        {"content": "Implement feature X", "activeForm": "Implementing feature X", "status": "completed"},
        {"content": "Write tests", "activeForm": "Writing tests", "status": "completed"},
        {"content": "Update docs", "activeForm": "Updating docs", "status": "in_progress"}
    ]
)
```

**When NOT to Use:**
- Operations depend on each other
- Need results from one to inform another
- Writing to same file

### Pattern 2: Progressive Search

Start broad, then narrow down:

```python
# Step 1: Find files with matches
Grep(
    pattern="class \\w+",
    type="py",
    output_mode="files_with_matches"
)
# Result: 150 files found

# Step 2: Narrow to specific pattern
Grep(
    pattern="class User\\w*",
    type="py",
    output_mode="files_with_matches"
)
# Result: 12 files found

# Step 3: Get actual content
Grep(
    pattern="class User\\w*",
    type="py",
    output_mode="content",
    **{"-A": 10}
)

# Step 4: Read specific files
Read(file_path="/home/user/project/src/models/user.py")
```

**Benefits:**
- Faster initial results
- Avoid overwhelming output
- Better understanding of codebase
- More targeted analysis

### Pattern 3: Read-Edit-Verify

Safe file modification workflow:

```python
# Step 1: Read current state
Read(file_path="/home/user/project/src/config.py")

# Step 2: Make precise edit
Edit(
    file_path="/home/user/project/src/config.py",
    old_string="DEBUG = True",
    new_string="DEBUG = False"
)

# Step 3: Verify change
Grep(
    pattern="DEBUG",
    path="/home/user/project/src/config.py",
    output_mode="content"
)

# Step 4: Test if critical
Bash(
    command="python -m pytest tests/test_config.py",
    description="Verify config changes don't break tests"
)
```

**Why This Matters:**
- Prevents accidental overwrites
- Confirms exact changes
- Catches errors early
- Maintains code quality

### Pattern 4: Batch File Operations

Process multiple files efficiently:

```python
# Step 1: Find target files
Glob(pattern="**/*test*.py")
# Result: test_user.py, test_auth.py, test_api.py

# Step 2: Read all test files in parallel
Read(file_path="/home/user/project/tests/test_user.py")
Read(file_path="/home/user/project/tests/test_auth.py")
Read(file_path="/home/user/project/tests/test_api.py")

# Step 3: Edit each file
Edit(
    file_path="/home/user/project/tests/test_user.py",
    old_string="import unittest",
    new_string="import pytest"
)
Edit(
    file_path="/home/user/project/tests/test_auth.py",
    old_string="import unittest",
    new_string="import pytest"
)
Edit(
    file_path="/home/user/project/tests/test_api.py",
    old_string="import unittest",
    new_string="import pytest"
)

# Step 4: Verify all changes
Bash(
    command="pytest tests/",
    description="Run all tests to verify migration to pytest"
)
```

Combine tools with git operations:

## Tool Permissions and Security

### File Access

**Read Tool:**
- Can access any file on the system
- Respects file permissions
- Safe for reading sensitive files
- No modification capability

**Write Tool:**
- Can create/overwrite files
- Requires confirmation for existing files
- Creates parent directories if needed
- Cannot delete files

**Edit Tool:**
- Can modify existing files
- Must read first (safety check)
- Exact string matching prevents accidents
- Cannot create new files

### Command Execution

**Bash Tool:**
- Runs with your user permissions
- Can execute any shell command
- No sudo prompts (would fail)
- Review commands before running

**Security Best Practices:**
1. Review generated commands
2. Don't commit secrets
3. Use environment variables for sensitive data
4. Be cautious with `rm`, `mv`, `chmod`
5. Validate file paths
6. Use hooks for security checks

### Dangerous Operations

**High Risk:**
```bash
# DON'T: Delete everything
rm -rf /

# DON'T: Force push to main
git push --force origin main

# DON'T: Execute untrusted code
curl http://untrusted.com/script.sh | bash

# DON'T: Change system permissions
chmod 777 -R /
```

**Safe Alternatives:**
```bash
# DO: Delete specific directory
rm -rf ./build

# DO: Regular push
git push origin feature-branch

# DO: Review before executing
curl http://trusted.com/script.sh > script.sh
# Review script.sh
bash script.sh

# DO: Specific permissions
chmod 755 ./scripts/build.sh
```

### Workflow 3: Debug Production Issue

## Best Practices for Tool Usage

### 1. Always Read Before Writing

```python
# ❌ BAD: Write without reading
Write(
    file_path="/home/user/project/config.py",
    content="NEW_CONFIG = True"
)

# ✅ GOOD: Read first, then edit
Read(file_path="/home/user/project/config.py")
Edit(
    file_path="/home/user/project/config.py",
    old_string="OLD_CONFIG = False",
    new_string="NEW_CONFIG = True"
)
```

### 2. Use Appropriate Search Tools

```python
# ❌ BAD: Using bash for search
Bash(command="find . -name '*.py' | xargs grep 'TODO'")

# ✅ GOOD: Use Glob and Grep
Grep(pattern="TODO", type="py", output_mode="content")
```

### 3. Prefer Edit Over Write

```python
# ❌ BAD: Overwrite entire file for small change
Read(file_path="/home/user/project/app.py")
# ... read 500 lines
Write(
    file_path="/home/user/project/app.py",
    content="... all 500 lines with one change ..."
)

# ✅ GOOD: Make targeted edit
Edit(
    file_path="/home/user/project/app.py",
    old_string="version = '1.0.0'",
    new_string="version = '1.0.1'"
)
```

### 4. Run Tests After Changes

```python
# Make changes
Edit(file_path="/home/user/project/src/feature.py", ...)

# ✅ GOOD: Verify with tests
Bash(
    command="pytest tests/test_feature.py",
    description="Verify changes"
)
```

### 5. Use Absolute Paths

```python
# ❌ BAD: Relative path
Read(file_path="./src/app.py")

# ✅ GOOD: Absolute path
Read(file_path="/home/user/project/src/app.py")
```

### 6. Provide Clear Descriptions

```python
# ❌ BAD: No description
Bash(command="npm test")

# ✅ GOOD: Clear description
Bash(
    command="npm test",
    description="Run test suite to verify changes"
)
```

### 7. Handle Large Files Efficiently

```python
# For very large files
Read(
    file_path="/home/user/project/large.log",
    offset=0,
    limit=100
)

# Or use Grep to find specific sections
Grep(
    pattern="ERROR",
    path="/home/user/project/large.log",
    output_mode="content",
    head_limit=50
)
```

### 8. Leverage Parallel Execution

```python
# ✅ GOOD: Read multiple files at once
Read(file_path="/home/user/project/src/app.py")
Read(file_path="/home/user/project/src/utils.py")
Read(file_path="/home/user/project/tests/test_app.py")
```

### 9. Be Specific with Grep Patterns

```python
# ❌ BAD: Too broad
Grep(pattern="user")

# ✅ GOOD: Specific pattern
Grep(
    pattern="class User[A-Z]\\w*",
    type="py",
    output_mode="content"
)
```

### 10. Maintain Exact Indentation

```python
# When editing, preserve exact indentation from Read output
Read(file_path="/home/user/project/app.py")
# Shows: "    def method():"  (4 spaces)

# ✅ GOOD: Match exactly
Edit(
    file_path="/home/user/project/app.py",
    old_string="    def method():",  # Exact 4 spaces
    new_string="    def new_method():"
)
```

```python
# ❌ Too generic
Edit(
    file_path="/home/user/project/app.py",
    old_string="return result",  # Appears 10 times!
    new_string="return processed_result"
)

# ✅ Fix 1: Add context
Edit(
    file_path="/home/user/project/app.py",
    old_string="""def process_data(data):
    result = transform(data)
    return result""",
    new_string="""def process_data(data):
    result = transform(data)
    return processed_result"""
)

# ✅ Fix 2: Use replace_all
Edit(
    file_path="/home/user/project/app.py",
    old_string="return result",
    new_string="return processed_result",
    replace_all=True
)
```

**Problem: Indentation mismatch**

### Grep Tool Issues

**Problem: No results found**

```python
# Check 1: Pattern might need escaping
# ❌ Wrong: Literal dot treated as "any character"
Grep(pattern="console.log")

# ✅ Fix: Escape special characters
Grep(pattern="console\\.log")

# Check 2: Case sensitivity
# ✅ Use -i for case-insensitive
Grep(pattern="TODO", **{"-i": True})

# Check 3: File type filter too restrictive
# ❌ Might miss files
Grep(pattern="function", type="js")

# ✅ Broaden search
Grep(pattern="function", glob="*.{js,jsx,ts,tsx}")
```

**Problem: Too many results**

```python
# ✅ Limit results
Grep(
    pattern="import",
    output_mode="content",
    head_limit=50
)

# ✅ Use more specific pattern
Grep(pattern="import \\w+ from 'react'")

# ✅ Filter by file type
Grep(pattern="import", type="js", glob="src/**")
```

**Problem: Multiline pattern not matching**

```python
# ❌ Without multiline mode
Grep(pattern="class \\w+[\\s\\S]*?\\{")

# ✅ Enable multiline
Grep(
    pattern="class \\w+[\\s\\S]*?\\{",
    multiline=True,
    output_mode="content"
)
```

## Tool Selection Decision Tree

**Need to find files?**
- By name/pattern → `Glob`
- By content → `Grep` with `files_with_matches`

**Need to search code?**
- Just file names → `Grep` with `files_with_matches`
- See matches → `Grep` with `content`
- Count matches → `Grep` with `count`

**Need to view a file?**
- Known path → `Read`
- Large file → `Read` with `offset`/`limit`
- Specific section → `Grep` with context

**Need to modify a file?**
- New file → `Write`
- Edit existing → `Edit`
- Notebook cell → `NotebookEdit`

**Need to run commands?**
- Quick command → `Bash`
- Long-running → `Bash` with `run_in_background`
- Monitor output → `BashOutput`
- Stop process → `KillShell`

**Need to track tasks?**
- Multi-step project → `TodoWrite`
- Single task → Just do it

**Need web content?**
- Specific URL → `WebFetch`
- Search internet → `WebSearch`
