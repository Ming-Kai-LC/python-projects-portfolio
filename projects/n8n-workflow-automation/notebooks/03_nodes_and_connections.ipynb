## IF Node: Conditional Branching

**Use cases:**
- Filter data based on conditions
- Route to different actions
- Validate data

### Example: Age Validation

```
Condition: {{$json.age}} >= 18
True path: Send welcome email
False path: Send parent consent form
```

````python
# Python equivalent of IF node logic
users = [
    {"name": "Alice", "age": 25},
    {"name": "Bob", "age": 16},
    {"name": "Charlie", "age": 30}
]

adults = [u for u in users if u['age'] >= 18]
minors = [u for u in users if u['age'] < 18]

print(f"Adults: {[u['name'] for u in adults]}")
print(f"Minors: {[u['name'] for u in minors]}")
````

## Switch Node: Multiple Paths

**Use when:** More than 2 conditions

### Example: Order Status Routing

```
Route 0: status === 'pending' → Send reminder
Route 1: status === 'shipped' → Send tracking
Route 2: status === 'delivered' → Request review
Route 3: status === 'cancelled' → Process refund
```

## Merge Node: Combining Data

**Modes:**
1. **Append**: Combine all items
2. **Merge By Key**: Join like SQL JOIN
3. **Wait**: Wait for all paths
