```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

# Enable inline plotting
%matplotlib inline

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. Matplotlib Basics

```python
# Simple line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
plt.show()
```

```python
# Multiple plots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Line plot
axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title('Sine')

# Scatter plot
axes[0, 1].scatter(x, np.cos(x))
axes[0, 1].set_title('Cosine')

# Bar plot
axes[1, 0].bar(['A', 'B', 'C', 'D'], [3, 7, 2, 5])
axes[1, 0].set_title('Bar Chart')

# Histogram
data = np.random.randn(1000)
axes[1, 1].hist(data, bins=30)
axes[1, 1].set_title('Histogram')

plt.tight_layout()
plt.show()
```

## 2. Seaborn for Statistical Plots

```python
# Create sample data
df = pd.DataFrame({
    'x': np.random.randn(100),
    'y': np.random.randn(100),
    'category': np.random.choice(['A', 'B', 'C'], 100)
})
```

```python
# Scatter plot with categories
sns.scatterplot(data=df, x='x', y='y', hue='category', s=100)
plt.title('Scatter Plot with Categories')
plt.show()
```

```python
# Box plot
sns.boxplot(data=df, x='category', y='x')
plt.title('Box Plot by Category')
plt.show()
```

```python
# Correlation heatmap
correlation = df[['x', 'y']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

## 3. Interactive Plots with Plotly

```python
# Interactive scatter plot
fig = px.scatter(df, x='x', y='y', color='category', 
                 title='Interactive Scatter Plot',
                 hover_data=['x', 'y'])
fig.show()
```

```python
# Interactive line plot
time_series = pd.DataFrame({
    'date': pd.date_range('2024-01-01', periods=100),
    'value': np.cumsum(np.random.randn(100))
})

fig = px.line(time_series, x='date', y='value', 
              title='Time Series Plot')
fig.show()
```

```python
# 3D Scatter plot
df_3d = pd.DataFrame({
    'x': np.random.randn(200),
    'y': np.random.randn(200),
    'z': np.random.randn(200),
    'category': np.random.choice(['A', 'B'], 200)
})

fig = px.scatter_3d(df_3d, x='x', y='y', z='z', color='category',
                    title='3D Scatter Plot')
fig.show()
```

## 4. Customization and Styling

```python
# Custom styled plot
plt.figure(figsize=(10, 6))
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), label='sin(x)', linewidth=2, color='#FF6B6B')
plt.plot(x, np.cos(x), label='cos(x)', linewidth=2, color='#4ECDC4', linestyle='--')

plt.title('Trigonometric Functions', fontsize=16, fontweight='bold')
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.legend(loc='upper right', fontsize=11)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='black', linewidth=0.5)
plt.axvline(x=0, color='black', linewidth=0.5)
plt.show()
```

## 5. Common Plot Types Quick Reference

| Plot Type | Matplotlib | Seaborn | Plotly |
|-----------|------------|---------|--------|
| Line | `plt.plot()` | - | `px.line()` |
| Scatter | `plt.scatter()` | `sns.scatterplot()` | `px.scatter()` |
| Bar | `plt.bar()` | `sns.barplot()` | `px.bar()` |
| Histogram | `plt.hist()` | `sns.histplot()` | `px.histogram()` |
| Box | `plt.boxplot()` | `sns.boxplot()` | `px.box()` |
| Heatmap | - | `sns.heatmap()` | `px.imshow()` |
