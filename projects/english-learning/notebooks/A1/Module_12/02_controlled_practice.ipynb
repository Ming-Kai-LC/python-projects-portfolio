### Key Points to Remember:
- **CAN** expresses ability and permission
- **CAN'T** expresses inability
- CAN never changes form (no -s, no -ed)
- Use base verb after CAN (no 'to')
- Questions: **CAN + subject + verb?**
- Don't use DO/DOES with CAN
- Short answers: **Yes, I can** / **No, I can't**
