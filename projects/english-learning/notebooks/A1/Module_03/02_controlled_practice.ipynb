```python
# Exercise 6.2
exercise.create_multiple_choice(
    "My brother ___ at a university.",
    {"A": "study", "B": "studys", "C": "studies"},
    "C",
    "Consonant + y becomes -ies."
)
```

```python
# Exercise 6.3
exercise.create_multiple_choice(
    "They ___ coffee every morning.",
    {"A": "drink", "B": "drinks", "C": "drinkes"},
    "A",
    "'They' uses the base form (no -s)."
)
```
