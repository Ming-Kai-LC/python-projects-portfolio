### Key Points to Remember:
- **WHAT** = things, information
- **WHERE** = places, locations
- **WHO** = people
- **WHEN** = time
- **WHY** = reasons (answer with BECAUSE)
- **HOW** = method, condition
- Word order: **Wh-word + BE/DO/DOES + Subject**
- Exception: WHO as subject doesn't need DO/DOES
