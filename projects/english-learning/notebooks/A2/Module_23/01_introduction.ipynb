### What You'll Learn

1. Use 'who' for people in relative clauses
2. Use 'which' and 'that' for things in relative clauses
3. Use 'where' for places in relative clauses
4. Form sentences with relative clauses correctly
5. Understand defining relative clauses (no commas at A2)
